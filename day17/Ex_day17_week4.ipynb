```python
def count_negative_numbers(numbers):
    negatives = [n for n in numbers if n < 0]
    return len(negatives)


nums = [3, -1, 0, -7, 5, -2]
print(count_negative_numbers(nums))
```

```python
import numpy as np


arr = np.random.randint(-50, 51, size=20)
filtered = arr[(arr >= -10) & (arr <= 10)]
mean_value = filtered.mean() if filtered.size > 0 else 0

abs_max = np.max(np.abs(arr))

print("Array:", arr)
print("Filtered (-10 to 10):", filtered)
print("Mean:", mean_value)
print("Absolute max:", abs_max)
```

```python
import numpy as np
import matplotlib.pyplot as plt


x = np.linspace(-30, 30, 200)
noise = np.random.randint(-20, 21, size=len(x))
y_noisy = x**2 - 4*x + noise
y_true = x**2 - 4*x


plt.figure()
plt.scatter(x, y_noisy, label="Noisy data")
plt.plot(x, y_true, label="True function")
plt.title("Noisy Data vs True Curve")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score


np.random.seed(42)
n = 500

age = np.random.randint(18, 66, size=n)
income = np.random.randint(30000, 150001, size=n)

X = np.column_stack([age, income])
y = ((age < 35) & (income > 80000)).astype(int)

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.30, random_state=42, stratify=y
)


model = LogisticRegression(max_iter=1000)
model.fit(X_train, y_train)


y_pred = model.predict(X_test)
acc = accuracy_score(y_test, y_pred)
print("Accuracy:", acc)


new_age = 28
new_income = 95000
new_pred = model.predict([[new_age, new_income]])[0]
new_prob = model.predict_proba([[new_age, new_income]])[0, 1]

print(f"New input: age={new_age}, income={new_income}")
print("Predicted class:", new_pred)
print("Prob(class=1):", new_prob)
```
