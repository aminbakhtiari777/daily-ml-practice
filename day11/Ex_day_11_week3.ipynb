```python
def max_difference(nums):
    if len(nums) < 2:
        return 0
    return max(nums) - min(nums)


print(max_difference([2, 9, 4, 15]))
```

```python
import numpy as np

arr = np.random.rand(4, 4)
row_means = arr.mean(axis=1)
max_value = arr.max()


diag_sum = np.trace(arr)

print("Array:\n", arr)
print("Row means:", row_means)
print("Max value:", max_value)
print("Diagonal sum:", diag_sum)
```

```python
import pandas as pd


df = pd.DataFrame({
    "age": [25, 32, 41, 29, 37],
    "salary": [12_000, 18_500, 24_000, 15_000, 28_000],
    "bought": ["yes", "no", "yes", "no", "yes"]
})

print("DataFrame:")
print(df)

mean_salary_over_30 = df[df["age"] > 30]["salary"].mean()
yes_count = (df["bought"] == "yes").sum()
df["bought_binary"] = df["bought"].map({"yes": 1, "no": 0})
print("\nMean salary (age > 30):", mean_salary_over_30)
print("Count of 'yes':", yes_count)
print("\nDataFrame with binary column:")
print(df)
```

```python
import numpy as np
import matplotlib.pyplot as plt


x = np.linspace(-5, 5, 200)
y = x**3 - 4*x


plt.figure(figsize=(6,4))
plt.plot(x, y)
plt.title("y = x^3 - 4x")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)

plt.show()
```

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score


data = {
    "age":    [22, 25, 27, 30, 35, 40, 45, 29, 33, 38],
    "income": [3000, 4000, 5000, 6000, 8000, 12000, 15000, 7000, 9000, 11000],
    "bought": [0,    0,    0,    1,    1,    1,     1,     0,    1,     1]
}

df = pd.DataFrame(data)
X = df[["age", "income"]]
y = df["bought"]


X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)


model = LogisticRegression(solver="liblinear")
model.fit(X_train, y_train)


y_pred = model.predict(X_test)
acc = accuracy_score(y_test, y_pred)
print("Accuracy:", acc)


new_sample = np.array([[32, 8500]])
pred = model.predict(new_sample)[0]
print("Prediction for (age=32, income=8500):", "Buy" if pred == 1 else "Not Buy")

```
