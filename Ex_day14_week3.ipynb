```python
def count_upper_lower(text):
    upper_count = 0
    lower_count = 0

    for ch in text:
        if ch.isupper():
            upper_count += 1
        elif ch.islower():
            lower_count += 1

    return upper_count, lower_count
s = "Hello World From AI"
upper, lower = count_upper_lower(s)
print("Uppercase:", upper)
print("Lowercase:", lower)
```

```python
import pandas as pd
import numpy as np


data = {
    "name": ["Ali", "Sara", "Reza", "Niloofar", "Mina"],
    "age": [25, 32, 28, 40, 35],
    "income": [42000, 68000, 51000, 90000, 47000],
    "score": [85, 90, 78, 88, 92]
}

df = pd.DataFrame(data)


df.loc[2, "score"] = np.nan


df["score"] = df["score"].fillna(df["score"].mean())


filtered_df = df[df["income"] > 50000]


print("Original & Cleaned DataFrame:\n", df)
print("\nFiltered (income > 50000):\n", filtered_df)
```

```python
import numpy as np
import matplotlib.pyplot as plt


x = np.linspace(-30, 30, 200)
noise = np.random.randint(-10, 11, size=len(x))
y_noisy = 0.5 * x**2 - 4 * x + noise
y_clean = 0.5 * x**2 - 4 * x


plt.scatter(x, y_noisy)
plt.plot(x, y_clean)
plt.title("Noisy Scatter vs Clean Curve")
plt.grid(True)
plt.show()
```

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score


np.random.seed(42)

n_samples = 300
feature1 = np.random.randint(18, 61, size=n_samples)
feature2 = np.random.randint(20000, 120001, size=n_samples)
target = np.where((feature1 < 30) & (feature2 > 80000), 1, 0)


df = pd.DataFrame({
    "age": feature1,
    "income": feature2,
    "target": target
})


X = df[["age", "income"]].values
y = df["target"].values


X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)

model = LogisticRegression()
model.fit(X_train, y_train)


y_pred = model.predict(X_test)
acc = accuracy_score(y_test, y_pred)
print("Accuracy:", acc)


new_sample = np.array([[25, 90000]])
pred_class = model.predict(new_sample)
pred_proba = model.predict_proba(new_sample)

print("New sample prediction (age=25, income=90000):", pred_class[0])
print("Class probabilities [class 0, class 1]:", pred_proba[0])
```
