```python
def returner(entry):
  avg=sum(entry)%len(entry)
  min_=min(entry)
  max_=max(entry)
  return avg,min_,max_


print(returner([1,3,4]))

```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-30, 30, 200)
noise = np.random.randint(-20, 21, len(x))

y_noisy = 0.5 * x**2 - 3*x + noise
y_clean = 0.5 * x**2 - 3*x

plt.figure(figsize=(8,5))

plt.scatter(x, y_noisy, label="Noisy Data", alpha=0.6)
plt.plot(x, y_clean, label="Clean Curve", linewidth=2)

plt.title("Scatter and Clean Curve")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()

plt.savefig("day_plot.png", dpi=300)
plt.show()
```

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

x = np.arange(1, 201).reshape(-1, 1)
noise = np.random.randint(-10, 11, x.shape[0])
y = 4 * x.flatten() + 12 + noise

X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

model = LinearRegression()
model.fit(X_train, y_train)

y_pred = model.predict(X_test)

mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print("MSE:", mse)
print("R2:", r2)

y_150 = model.predict([[150]])
print("Prediction for x=150:", y_150[0])

```
