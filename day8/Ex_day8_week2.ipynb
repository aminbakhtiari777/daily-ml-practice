```python
import pandas as pd
import numpy as np

names = ["Amin", "Shiva", "Nima", "Sara", "Reza"]

df = pd.DataFrame({
    "Name": names,
    "Age": np.random.randint(18, 50, size=5),
    "Score": np.random.randint(50, 100, size=5)
})
mean_score=df["Score"].mean()
print(mean_score)
print(df.head(2))
print(df.info())
print(df.describe())
print(df.shape)
```

```python
import matplotlib.pyplot as plt
import numpy as np


x = np.linspace(-10, 10, 200)
y = 2*x + 5

plt.plot(x, y)
plt.title("Simple Linear Function: y = 2x + 5")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)

plt.show()
```

```python
import numpy as np
from sklearn.linear_model import LinearRegression


x = np.arange(1, 101).reshape(-1, 1)
noise = np.random.randint(-5, 6, size=(100, 1))
y = 3*x + 7 + noise

model = LinearRegression()
model.fit(x, y)


y_pred = model.predict([[50]])

print("Prediction for x = 50:", y_pred[0][0])
```
