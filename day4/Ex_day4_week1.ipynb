```python
import numpy as np
def sum_(numbers):
  even_sum = 0
  odd_sum = 0
  for n in numbers:
    if n%2==0:
      even_sum+=n
    else:
      odd_sum+=n
  return {"even_sum":even_sum,
          "odd_sum":odd_sum}

nums = [1, 2, 3, 4, 5, 6]
result = sum_(nums)
print(result)

```

```python
import numpy as np
matrix=np.random.randint(1,10,size=(4,4))
main_diag=np.diag(matrix)
print(main_diag)
sec_diag=np.diag(np.fliplr(matrix))
print(sec_diag)
row_sum=np.sum(matrix,axis=0)
print(row_sum)
```

```python
import pandas as pd
import numpy as np

data = {
    "Age": np.random.randint(18, 61, size=10),
    "Score": [45, "A", 60, 30, "?", 80, 55, 100, "B", 70]
}
df = pd.DataFrame(data)
print("Original DataFrame:")

df["Score"]=pd.to_numeric(df["Score"],errors="coerce")

df_clean=df.dropna(subset=["Score"])
print(df_clean)

mean_age=df_clean["Age"].mean()
print(mean_age)

df_high_score=df_clean[df_clean["Score"]>50]
print(df_high_score)
```

```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

x = np.arange(0, 51, 1)
noise = np.random.randint(-5, 6, size=len(x))
y = 5 * x + 12 + noise

x = x.reshape(-1, 1)

x_train, x_test, y_train, y_test = train_test_split(
    x, y, test_size=0.2, random_state=42
)

model = LinearRegression()
model.fit(x_train, y_train)


x_new = np.array([[20]])
y_pred_20 = model.predict(x_new)
print("Predicted y for x=20:", y_pred_20[0])
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(-10, 10.5, 0.5)
noise = np.random.uniform(-3, 3, size=len(x))
y_noisy = 2 * x**2 - 4 * x + 3 + noise
y_clean = 2 * x**2 - 4 * x + 3


plt.figure()
plt.scatter(x, y_noisy, label="Noisy data")
plt.plot(x, y_clean, label="True function")
plt.title("Quadratic Function with Noise")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.savefig("day_plot.png", dpi=300)
plt.show()
```
