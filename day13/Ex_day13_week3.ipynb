```python
def count_persian_vowels(text):
    vowels = "اآوی"
    count = 0
    for ch in text:
        if ch in vowels:
            count += 1
    return count

# Test
text = "امین امروز کدنویسی کرد"
result = count_persian_vowels(text)
print(result)


```

```python
import pandas as pd
import numpy as np


data = {
    "name": ["Ali", "Sara", "Reza", "Mina", "Omid"],
    "age": [25, 34, 29, 41, 37],
    "income": [4000, 7200, 5100, 8800, 6400],
    "score": [85, 90, np.nan, 78, 88]
}

df = pd.DataFrame(data)
df["score"] = df["score"].fillna(df["score"].mean())
mean_income_over_30 = df[df["age"] > 30]["income"].mean()

print(df)
print("\nMean income of people over 30:", mean_income_over_30)
```

```python
import numpy as np
import matplotlib.pyplot as plt


x = np.linspace(-20, 20, 400)
y1 = x**2
y2 = 3*x + 1


plt.figure()
plt.subplot(2, 1, 1)
plt.plot(x, y1)
plt.title("y = x^2")
plt.grid(True)
plt.subplot(2, 1, 2)
plt.plot(x, y2)
plt.title("y = 3x + 1")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score


np.random.seed(42)

feature1 = np.random.randint(1, 101, 200)
feature2 = np.random.rand(200)
target = np.where((feature1 > 60) & (feature2 > 0.4), 1, 0)
X = np.column_stack((feature1, feature2))
y = target


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)


model = RandomForestClassifier()
model.fit(X_train, y_train)


y_pred = model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)

print("Accuracy:", accuracy)
new_input = np.array([[80, 0.3]])
prediction = model.predict(new_input)
print("Prediction for (80, 0.3):", prediction[0])
```
