```python
import numpy as np

data = np.random.randint(0, 101, size=100)
mean_all = data.mean()
subset = data[data > mean_all]


subset_mean = subset.mean()
subset_std = subset.std()

print("Mean of all data:", mean_all)
print("Subset mean:", subset_mean)
print("Subset standard deviation:", subset_std)
```

```python
import pandas as pd
import numpy as np


df = pd.DataFrame({
    "age": [25, 32, np.nan, 45, np.nan, 29, 41],
    "income": [50000, 72000, 61000, 90000, 58000, 67000, 83000],
    "city": [" Tehran ", "tehran", "Shiraz", " shiraz ", "TEHRAN", "Isfahan ", " isfahan"]
})


age_median = df["age"].median()
df["age"] = df["age"].fillna(age_median)


df["city"] = df["city"].str.lower().str.strip()


city_counts = df["city"].value_counts()

print(df)
print("\nCity counts after cleaning:")
print(city_counts)
```

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score


np.random.seed(42)

n = 1000

age = np.random.randint(18, 66, n)
income = np.random.randint(30000, 150001, n)
visits = np.random.randint(1, 11, n)


target = ((income > 90000) & (visits > 6)).astype(int)

df = pd.DataFrame({
    "age": age,
    "income": income,
    "visits": visits,
    "target": target
})

X = df[["age", "income", "visits"]]
y = df["target"]


X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)


model = LogisticRegression(max_iter=1000)
model.fit(X_train, y_train)


y_pred = model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)

print("Accuracy:", accuracy)
```

```python
# If the model has high training accuracy but low test accuracy, it is suffering from overfitting (high variance), meaning it learns noise and specific patterns of the training data rather than generalizable relationships.

# A simple way to reduce this is regularization or reducing model complexity (alternatively, collecting more training data).

# Adding more features is not always beneficial because irrelevant or noisy features can increase variance, making the model fit noise instead of true signal and harming generalization.
```
