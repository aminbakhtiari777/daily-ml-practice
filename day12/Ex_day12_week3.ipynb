```python
def sum_even_numbers(numbers):
    total = 0
    for num in numbers:
        if num % 2 == 0:
            total += num
    return total



nums = [3, 10, 8, 5, 14, 7]
result = sum_even_numbers(nums)
print(result)
```

```python
import numpy as np

matrix = np.random.randint(0, 11, size=(3, 3))
print("Matrix:")
print(matrix)

det = np.linalg.det(matrix)
print("\nDeterminant:")
print(det)

transpose = matrix.T
print("\nTranspose:")
print(transpose)

if det != 0:
    inverse = np.linalg.inv(matrix)
    print("\nInverse:")
    print(inverse)
else:
    print("\nInverse does not exist (determinant is zero).")
```

```python
import pandas as pd
import numpy as np


df = pd.DataFrame({
    "name": ["Ali", "Sara", "Reza", "Mina", "Hossein"],
    "age": [28, 35, 32, 26, 40],
    "salary": [5000, 8000, 7500, 4500, 9000],
    "score": [85, 90, 88, 92, 80]
})

print("Original DataFrame:")
print(df)


df.loc[1, "score"] = np.nan
df.loc[4, "score"] = np.nan

print("\nWith NaN values:")
print(df)


mean_score = df["score"].mean()
df["score"] = df["score"].fillna(mean_score)

print("\nAfter Cleaning NaN with Mean:")
print(df)

mean_salary_above_30 = df[df["age"] > 30]["salary"].mean()

print("\nMean salary of people older than 30:")
print(mean_salary_above_30)
```

```python
import numpy as np
import matplotlib.pyplot as plt


x = np.linspace(-10, 10, 200)
y = x**2 - 5*x + 6


plt.figure()
plt.plot(x, y, color="purple")
plt.grid(True, alpha=0.3)
plt.xlabel("x")
plt.ylabel("y")
plt.title("y = x^2 - 5x + 6")

plt.show()
```

```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score


np.random.seed(42)
feature1 = np.random.randint(1, 101, 200)
feature2 = np.random.rand(200)
X = np.column_stack((feature1, feature2))
y = np.where((feature1 > 50) & (feature2 > 0.5), 1, 0)


X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)


model = DecisionTreeClassifier()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)


accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)

new_input = np.array([[80, 0.3]])
prediction = model.predict(new_input)

print("New input (80, 0.3) prediction:", prediction[0])
```
