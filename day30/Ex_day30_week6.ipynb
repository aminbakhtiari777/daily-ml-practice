```python
import numpy as np
import pandas as pd

np.random.seed(0)

df = pd.DataFrame({
    "age": np.random.randint(18, 71, 1000),
    "income": np.random.randint(30000, 150001, 1000),
    "city": np.random.choice(["A", "B", "C"], 1000)
})

mean_all = df["income"].mean()
mean_A = df[df["city"] == "A"]["income"].mean()

print(mean_all, mean_A)
```

```python
import matplotlib.pyplot as plt


df.groupby("city")["income"].mean().plot(kind="bar")
plt.xlabel("City")
plt.ylabel("Mean Income")
plt.show()


df["age_group"] = np.where(df["age"] < 40, "<40", ">=40")

pivot = df.groupby(["age_group", "city"])["income"].mean().unstack()
pivot.plot(kind="bar")
plt.xlabel("Age Group")
plt.ylabel("Mean Income")
plt.show()
```

```python
# If features are not scaled, income (with a much larger numerical range) will dominate the distance calculation, and education_years will have very little influence. Since KNN relies on distance, the model will effectively make decisions based almost entirely on income, leading to biased and misleading neighbors.
```

```python
# When k is very small, KNN becomes highly sensitive to noise and individual data points, resulting in high variance and overfitting.
# When k is very large, the model averages over many points, producing an overly smooth decision boundary with high bias and underfitting.
```
