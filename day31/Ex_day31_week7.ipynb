```python
def keep_positives(numbers):
    positives = []
    for n in numbers:
        if n >= 0:
            positives.append(n)
    return positives


data = [3, -2, 10, -5, 0, 8]
result = keep_positives(data)
print(result)
```

```python
import numpy as np


matrix = np.random.rand(4, 4)
print("Matrix:\n", matrix)

mean_value = np.mean(matrix)
print("\nMean of matrix:", mean_value)

max_value = np.max(matrix)
print("Max value:", max_value)


normalized_matrix = matrix - mean_value
print("\nMatrix after subtracting mean:\n", normalized_matrix)
```

```python
import pandas as pd
import numpy as np


data = {
    "name": ["Ali", "Sara", "Reza", "Mina", "Hamed", "Neda"],
    "age": [25, 30, 22, 28, 35, 27],
    "city": ["Tehran", "Shiraz", "Tehran", "Tabriz", "Shiraz", "Tehran"],
    "score": [85, 90, 78, np.nan, 88, 92]
}

df = pd.DataFrame(data)
print("Initial DataFrame:\n", df)


mean_score = df["score"].mean()
df["score"] = df["score"].fillna(mean_score)

print("\nAfter filling NaN with mean:\n", df)

grouped = df.groupby("city")["score"].mean()
print("\nMean score by city:\n", grouped)
```

```python
import numpy as np
import matplotlib.pyplot as plt


x = np.linspace(-40, 40, 200)
noise = np.random.uniform(-15, 15, size=len(x))
y_noisy = 3 * x + noise
y_true = 3 * x

plt.scatter(x, y_noisy)
plt.plot(x, y_true)
plt.title("Noisy Scatter vs True Regression Line (y = 3x)")
plt.grid(True)
plt.show()
```

```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

np.random.seed(42)
n_samples = 200
feature1 = np.random.randint(18, 61, size=n_samples)
feature2 = np.random.randint(20000, 120001, size=n_samples)
target = np.where((feature1 < 35) & (feature2 > 70000), 1, 0)


X = np.column_stack((feature1, feature2))
y = target


X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)


knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, y_train)


y_pred = knn.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)


new_sample = np.array([[28, 90000]])
new_pred = knn.predict(new_sample)
print("Prediction for (age=28, income=90000):", new_pred[0])
```
